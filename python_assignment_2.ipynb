# **Sets (Operations)**

a. Create a set called my_set with following values:    
['a','e','i','o','u','a','a','i']
Analyse the output and provide explanation for the same

```python
my_set=set( ['a','e','i','o','u','a','a','i'] )
print(my_set)
```

# **Explanations:**
1.A set is a collection that stores only unique elements.

2.Duplicate values are automatically removed.

3.'a' appears 3 times
  'i' appears 2 times

4.Only unique vowels remain.

5.Order is not guaranteed (sets are unordered).

b. Attempt to change the value of my_set[4] = 's'. If code throws an error, provide
an explanation

```python
my_set[4]='s'
```

# **Explanations:**
1.Set is unordered → no indexing.

2.No duplicates → only unique values are stored.

3.Cannot use indexing like my_set[4].

4.TypeError occurs if you try to assign by index.

5.To update → use .remove() and .add() instead.

2. Find the performance category based on the input score using following criteria:
a. Above Average:  Score greater than 7
b. Average: Score between 4 and 7(both inclusive)
c. Below Average: Score lesser than 4

```python
score = int(input("Enter the score (0 to 10): "))

if score > 10 or score < 0:
    print("Enter a valid score (0 to 10)")
elif score > 7:
    print("Performance category: Above Average")
elif score >= 4:
    print("Performance category: Average")
else:
    print("Performance category: Below Average")
```
